```python
student={'Name':'Teerth',
       'Age':18,
       'Course':'BSc. Economics Hons.'}
print(student)
print(type(student))
print('='*45)

print(student.keys())
print(student.values())
print('='*45)

print(student['Name'])
print(student['Age'])
print(student['Course'])
print('='*45)

student['Age']=28
print(student)
print('='*45)

student['Hobby']='Coding'
print(student)
print('='*45)

student.pop('Course')
print(student)
```

```python
#GROCERY STORE
print('Welcome to the Grocery Store')
print('*'*45)
#SHOPKEEPER
store={'oil':60,
       'wheat':20,
       'milk':5}
print('Thats All I Got')
print(store)
print('='*45)
#CUSTOMER
print('What Do You Wanna Buy?')
item_to_buy=(input('Tell me What You Want?'))
qty=(int(input('Tell me How much You Want?')))
print('I want', item_to_buy, qty, 'Qty')
print('='*45)
#BILL
price=store[item_to_buy]
total=price*qty
print("Your Bill is:", total)
print('='*45)

store['eggs'] = 15
print('After adding eggs:', store)
print('=' * 45)

store.pop('eggs')
print('After removing eggs:', store)
```
