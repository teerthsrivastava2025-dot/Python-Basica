```python
import random #Choose a random number automatically
print("WELCOME TO GUESS WORLD")

secret_number = random.randint(1,15)
print('Secret Number guessed')
#terminate the game, if user guesses correctly or all 5 attempts are used up
for i in range(5):
    uno = int(input('Guess the number between 1-15:'))
    
    if uno == secret_number:
      print("COORECT GUESSS")
      break
    if uno > secret_number:
      print("TOO HIGH")
    if uno < secret_number:
      print("TOO LOW")

print('You Lost')
```
